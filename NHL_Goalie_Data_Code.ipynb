```python
#For loop to create dataframe of season stats
#testing it out with only 2 URLs

list_rows = []
season = []

#iterate over each URL
for url in urls:
    html = urlopen(url)
    soup = BeautifulSoup(html, 'lxml')
    rows = soup.find_all('tr')
    title = soup.title.get_text()[0:7]
    print(title, "finished scraping")

#iterate over each row to clean it up and add it to a list of the rows
    for row in rows:
        cells = row.find_all('td')
        str_cells = str(cells)
        clean = re.compile('<.*?>')
        clean2 = (re.sub(clean, '',str_cells))
        list_rows.append(clean2)
        season.append(title)
```

```python
df = pd.DataFrame(list_rows)
```

```python
df['Season']=season
```

```python
df.set_axis(['0','Season'], axis=1, inplace=False)
```

```python
i = df.columns.get_loc(0)
df2 = df[0].str.split(",", expand=True)
df2=pd.concat([df.iloc[:, :i], df2, df.iloc[:, i+1:]], axis=1)
```
